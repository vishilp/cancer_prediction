```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df.describe()
```

Now let's clean the data a bit

```python
df.isnull().sum() #find which null values to remove
```

```python
df.drop(['id','Unnamed: 32'], axis=1, inplace= True)
df.head()
```

```python
df['diagnosis'] = df['diagnosis'].apply(lambda x: 1 if x == 'M' else 0) #set malignant as 1 and others as 0
df['diagnosis'].value_counts()
```

```python
y = df['diagnosis']
X= df.drop(['diagnosis'], axis=1)
```

Normalize the data because the predictors are in much different scales

```python
from sklearn.preprocessing import StandardScaler
#use standardscaler because data should be normally distributed

#create a scaler object and fit scaler to data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled
```

Split the Data into Training and Testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42) 
```

Train the Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

#train model 
model.fit(X_train, y_train)
```

```python
#predict test data from trained model
y_pred = model.predict(X_test)
```

Evaluate model

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
print(f'Accuracy: {acc}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
